# Weight Initialization

why it is important ?

+ vanishing gradient problem ---> reason = weight initialization , sigmoid function
+ exploding gradient problem
+ slow convergence


## What not to do ...

* Case 1 : Zero initialization -- for relu and tanh weights will always be zero ; and for sigmoid model will act as linear model

* Case 2 : Non zero constant value -- model will act as a linear model, cannot capture non-linearity

* Case 3 : Random Initialization - small values (-0.01 , 0.01)
    + Vanishing Gradient problem for all functions

* Case 3.1 : Random initialization - large values (-3 , 3)
    + Relu : Exploding gradient problem
    + Sigmoid , Tanh : Vanishing gradient problem
