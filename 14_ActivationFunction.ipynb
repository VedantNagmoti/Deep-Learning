# Ideal Activation Function

+ Non linear Activation function
+ Diffrentiable
+ Computationally Inexpensive
+ Zero Centered / Normalized ---> Mean = 0 ---> Faster Convergence if normalized data is given Ex: Tanh
+ Non Saturating --> Ex : Relu - max(0 , x)
      
      for saturating function ex: sigmoid( 0 ,1 ), tanh(-1 , 1) Squeezes the input ---> Causing vanishing gradient problem


## Sigmoid Function

f(x) = 1 / 1 + e ^ -x

### Advantages
1. [0 , 1]   Probabillity ---> Output Layer Binary Classification
2. For non linear data --> Good option
3. Diffrentiable 

### Disadvantages
1. Saturating Function ---> [- , ] --> [0 , 1]    Vanishing gradient problem
2. Non zero centered 
3. if used in hidden layers it will take exponential time to converge




## Tanh Function

f(x) = e^x - e^-x / e^x + e^-x

### Advantages
+ non linear
+ diffrentiable
+ zero centered 
+ training faster than sigmoid 

### Disadvantages
+ Saturating Function - vanishing gradient problem
+ computationaly expensive 




## Relu Function

f(x) = max(0 , x)

### Advantages
+ Non linear
+ not saturated in the positive region
+ computationaly inexpensive
+ converge faster

### Disadvantages
+ differentiation   
                if x >= 0 ----> 1    
                if x < 0 ----> 0
+ Not zero centered 
+ Dying relu problem

# Dying relu problem

## Solutions

+ Set low learning rate
+ set bias -> positive value (0.01)
+ Dont use relu , use varients

## Leaky Relu

f(z) = max(0.01 * z , z)

+ if z >= 0  then  z
+ if z < 0   then  1/100 * z

### Advantages 

+ Non saturated - unbounded
+ easily computed
+ no dying relu problem 
+ close to zero centered
