# Batch Normalization

**What is batch normalization ?**

Batch Normalization (BatchNorm) is a technique for training deep neural networks that stabilizes learning and accelerates convergence by re-centering and scaling layer inputs. It normalizes the activation outputs of a previous layer to have a mean of zero and unit variance within each mini-batch, usually applied before the activation function.

## How to apply

+ Mini batch gradient descent 
+ layer by layer


## Advantages

+ Stable -> Hyperparameter -> Wider range of values 
+ training becomes faster 
+ can act as a regularizer - Not strong 
+ reduces impact of weight initialization
