```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
img_path = "CatDogClassification/test/cats/cat_1.jpg"
img = tf.keras.utils.load_img(img_path)
plt.imshow(img)
```

```python
x = img_to_array(img)
print(x)
```

```python
print(x.shape)
print(type(x))
```

```python
input_batch = x.reshape(1 , 769 , 615 , 3)
```

```python
print(input_batch)
```

```python
i = 0

for output in datagen.flow(input_batch , batch_size=1 , save_to_dir='Augmented'):
    i = i + 1
    if i == 10:
        break
```
