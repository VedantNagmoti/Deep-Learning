```python
model = Sequential()

model.add(Conv2D(32 , kernel_size=(3 , 3) , padding='valid' ,activation='relu' , input_shape = (28,28,1)))
model.add(Conv2D(32 , kernel_size=(3 , 3) , padding='valid' , activation='relu'))
model.add(Conv2D(32 , kernel_size=(3 , 3) , padding='valid' , activation='relu'))

model.add(Flatten())

model.add(Dense(128 , activation = 'relu'))
model.add(Dense(10 , activation = 'softmax'))
```

```python
model = Sequential()

model.add(Conv2D(32 , kernel_size=(3 , 3) , padding='same' ,activation='relu' , input_shape = (28,28,1)))
model.add(Conv2D(32 , kernel_size=(3 , 3) , padding='same' , activation='relu'))
model.add(Conv2D(32 , kernel_size=(3 , 3) , padding='same' , activation='relu'))

model.add(Flatten())

model.add(Dense(128 , activation = 'relu'))
model.add(Dense(10 , activation = 'softmax'))
```

```python
model.summary()
```

Padding - adding 0's on border so that after each convolution layer image size will remain same


stride - by how much step will kernel move through the image 

strided convolutions - stride more than 1 

###  why to use stride 

+ only care about high level features (less complex features) then use stride

+ reducing computational cost 

### Implementation of stride 

```python
model = Sequential()

model.add(Conv2D(32 , kernel_size=(3 , 3) , padding='same' , strides=(2 , 2) , activation='relu' , input_shape = (28,28,1)))
model.add(Conv2D(32 , kernel_size=(3 , 3) , padding='same' , strides=(2 , 2)  , activation='relu'))
model.add(Conv2D(32 , kernel_size=(3 , 3) , padding='same' , strides=(2 , 2)  , activation='relu'))

model.add(Flatten())

model.add(Dense(128 , activation = 'relu'))
model.add(Dense(10 , activation = 'softmax'))
```
