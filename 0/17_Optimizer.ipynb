# Exponentially weighted moving average

The Exponentially Weighted Moving Average (EWMA) is a statistical technique for modeling or smoothing time-series data that assigns exponentially decreasing weights to older observations while prioritizing recent data.

# Adagrad Optimizer

adaptive gradient -> learning rate is not fixed, we adapt it , change it 

Best works when :
1. **input features have different scales**
    + example : 
        + cgpa : 0 to 10
        + salary : 0 to lakhs
        + haveLoan : yes/no

    Generally we dont use this in this condition as we do normalization

2. **features are sparse -> most of the values are zero**
    + example : 
        | iq | cgpa | fromIIT | package |
        |----|------|---------|---------|
        |    |      |    0    |         |
        |    |      |    1    |         |
        |    |      |    0    |         |
    mostly the fromiit column is 0 because less students go to iit, but that column has so much significance for prediction of package column
    
    sparse features problems : 
    + elongated bowl problem

### Disadvantage
+ dont converge to solution, it just direct us near minima,
+ we mostly use for linear regression not neural network
